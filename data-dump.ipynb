```python
import os
import pandas as pd
from sqlalchemy import create_engine
import psycopg2

# Folder containing cleaned Olympic CSVs
csv_folder = './data'

# List of CSV files
csv_files = [
    "Athletes.csv",
    "Coaches.csv",
    "EntriesGender.csv",
    "Medals.csv",
    "Teams.csv"
]

# PostgreSQL database credentials
db_user = 'olympic_user'
db_password = 'olympic_pass'
db_host = 'localhost'
db_port = '5432'
db_name = 'olympics'

# Connect to PostgreSQL
connection = psycopg2.connect(
    database=db_name,
    user=db_user,
    password=db_password,
    host=db_host,
    port=db_port
)

# Create SQLAlchemy engine
engine = create_engine(f'postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')

def load_csv_to_postgresql(csv_file, table_name, conn, engine):
    try:
        df = pd.read_csv(csv_file, encoding='utf-8')
    except UnicodeDecodeError:
        df = pd.read_csv(csv_file, encoding='latin1')

    try:
        # Dynamically create table with all columns as TEXT
        columns = ', '.join([f'"{col}" TEXT' for col in df.columns])
        create_table_query = f"""
            CREATE TABLE IF NOT EXISTS "{table_name}" (
                {columns}
            );
        """

        print(f"\nüì• Creating table '{table_name}'...")
        with conn.cursor() as cursor:
            cursor.execute(create_table_query)
            conn.commit()

        print(f"üìä Inserting data into '{table_name}' ({df.shape[0]} rows)...")
        df.to_sql(table_name, engine, if_exists='replace', index=False, method='multi', chunksize=1000)
        print(f"‚úÖ Loaded {csv_file} into table '{table_name}'.")

    except Exception as e:
        print(f"‚ùå Error loading {csv_file}: {e}")

# Start uploading all CSVs
if not os.path.exists(csv_folder):
    print(f"‚ùå Folder '{csv_folder}' not found.")
else:
    for csv_file in csv_files:
        csv_file_path = os.path.join(csv_folder, csv_file.strip())
        table_name = os.path.splitext(csv_file)[0].lower()  # Lowercase table names
        load_csv_to_postgresql(csv_file=csv_file_path, table_name=table_name, conn=connection, engine=engine)

if connection:
    connection.close()
    print("üîí PostgreSQL connection closed.")
```
